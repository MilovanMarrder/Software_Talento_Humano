```python
import sqlite3
import pandas as pd

# 1. Conexión a la base de datos
conn = sqlite3.connect("rrhh.db")

# 2. Obtener la lista de tablas
query_tablas = "SELECT name FROM sqlite_master WHERE type='table';"
tablas = pd.read_sql_query(query_tablas, conn)['name']

# 3. Abrir el archivo de texto para escribir
with open("resumen_base_datos.txt", "w", encoding="utf-8") as f:
    f.write("=== VISIÓN GENERAL DE LA BASE DE DATOS RRHH ===\n\n")
    
    for tabla in tablas:
        # Consultar los datos de la tabla
        df = pd.read_sql_query(f"SELECT * FROM {tabla}", conn)
        total_registros = len(df)
        
        # Escribir encabezado de la tabla en el TXT
        f.write(f"{'='*50}\n")
        f.write(f"TABLA: {tabla}\n")
        f.write(f"Total de registros: {total_registros}\n")
        f.write(f"{'='*50}\n\n")
        
        if total_registros > 0:
            # Convertimos el DataFrame a string para que se vea como tabla
            f.write(df.to_string(index=False))
        else:
            f.write("Esta tabla está vacía.")
            
        f.write("\n\n") # Espacio entre tablas

# 4. Cerrar conexión
conn.close()

print("El archivo 'resumen_base_datos.txt' ha sido generado con éxito.")
```
